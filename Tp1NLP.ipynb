```python
#mise en minuscule le text
df['text_cleaned'] = df['text'].str.lower()
print(df['text_cleaned'] )
```

```python
#suprimer la ponctuation
import string
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
#afficher tout le text
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
# Définition de l'option pour afficher le contenu complet des cellules
pd.set_option('display.max_colwidth', None)
print(df['text_cleaned'])
```

```python
#supression des stopwords
from nltk.corpus import stopwords

# Téléchargez la liste des stopwords pour la première utilisation
nltk.download('stopwords')

# Chargez la liste des stopwords pour l'anglais
stop_words = set(stopwords.words('english'))

# Supprimez les mots vides de chaque ligne de votre DataFrame
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))

#afficher le text
print(df['text_cleaned'])
```

```python
#suprimer les mots frequents
from collections import Counter

# Comptage des mots fréquents
word_counts = Counter(' '.join(df['text_cleaned']).split())

# Nombre de mots fréquents à supprimer :les 10 mots les plus fréquents
num_common_words = 10
most_common_words = word_counts.most_common(num_common_words)

# Liste des mots fréquents à supprimer
common_words_to_remove = [word[0] for word in most_common_words]

# Suppression des mots fréquents
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in common_words_to_remove]))

#afficher le text
print(df['text_cleaned'])
```

```python
#supression des mots rares
# Tokenisation des mots dans la colonne "text"
mots = ' '.join(df['text_cleaned']).split()

# Compter les occurrences de chaque mot
occurrences_mots = Counter(mots)

# Seuil minimum d'occurrences pour considérer un mot comme rare
seuil_min_occurrences = 2

# Supprimer les mots rares
mots_communs = {mot for mot, occurrence in occurrences_mots.items() if occurrence >= seuil_min_occurrences}

# Fonction pour filtrer les mots rares
def filtrer_mots_rare(texte):
    mots_du_texte = texte.split()
    mots_filtrés = [mot for mot in mots_du_texte if mot in mots_communs]
    return ' '.join(mots_filtrés)

# Appliquer la fonction de filtrage sur la colonne "text"
df['text_cleaned'] = df['text_cleaned'].apply(filtrer_mots_rare)

# Afficher le DataFrame avec les mots rares supprimés
print(df['text_cleaned'])
```

```python
#stemming
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))
print(df['text_cleaned'])
```

```python
#Lemmatisation
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
print(df['text_cleaned'])
```

```python
#supression des emogies et emoticons
import re

def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# Appliquer la fonction remove_emojis sur la colonne "text_cleaned"
df['text_cleaned'] = df['text_cleaned'].apply(remove_emojis)

# Afficher le DataFrame avec les emojis supprimés
print(df['text_cleaned'])
```

```python
#supression des urls
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

df['text_cleaned'] = df['text_cleaned'].apply(remove_urls)
print(df['text_cleaned'])
```

```python
#supression des balises html
from bs4 import BeautifulSoup

def remove_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()
#appliquer la fonction remove_html_tags sur text_cleaned
df['text_cleaned'] = df['text_cleaned'].apply(remove_html_tags)
text_nv = df['text_cleaned']
print(df['text_cleaned'])
```
